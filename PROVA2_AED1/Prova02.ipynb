```python
def consumo(lista):
    total = sum(lista)
    media = total / len(lista)
    return media

consumo_diario = [50, 45, 55, 60, 48, 52, 47, 49, 53, 51, 50, 46, 54, 61, 49, 53, 48, 50, 52, 47, 56, 50, 55, 60, 48, 52, 47, 49, 53, 51]

print(f"O consumo diário médio foi: {consumo(consumo_diario)}")
```
